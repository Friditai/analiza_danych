```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
path = "./data/StressLevelDataset.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
unchanged_df = df.copy()
```

**Sprawdzenie korelacji**

```python
# Obliczanie korelacji miÄ™dzy rÃ³Å¼nymi cechami
correlation = unchanged_df.corr()
# Ustawienie wiÄ™kszego rozmiaru heatmapy
plt.figure(figsize=(15, 12))
# Tworzenie heatmapy korelacji
sns.heatmap(correlation, annot=True,  annot_kws={"size": 12})
# WyÅ›wietlenie wykresu
plt.show()
```

**Sprawdzenie unikatowych wartoÅ›ci**

```python
df['blood_pressure'].unique()
```

```python
blood_pressure_map = {1: 'low', 2: 'medium', 3: 'high'}
```

```python
df['blood_pressure_level'] = df['blood_pressure'].replace(blood_pressure_map)
```

```python
df[['blood_pressure','blood_pressure_level']].head(10)
```

```python
df['stress_level'].unique()
```

```python
stress_level_map = {0: 'low', 1: 'medium', 2: 'high'}
```

```python
df['stress_intensity'] = df['stress_level'].replace(stress_level_map)
```

```python
df[['stress_level','stress_intensity']].head(10)
```

```python
df.head()
```

**WpÅ‚yw poziomu odczuwanego stresu na jakoÅ›Ä‡ snu**

```python
# Tworzenie wykresu skrzypcowego pokazujÄ…cego rozkÅ‚ad i gÄ™stoÅ›Ä‡ poziomu odczuwanego stresu od jakoÅ›ci snu
sns.violinplot(x='stress_intensity', y='sleep_quality', data=df, order=['low', 'medium', 'high'])
```

Im mniejsza intensywnoÅ›Ä‡ przeÅ¼ywanego stresu tym lepsza jakoÅ›Ä‡ snu.

**WpÅ‚yw poziomu odczuwanego stresu na pojawienie siÄ™ depresji**

```python
sns.boxplot(x='stress_intensity', y='depression', data=df, order=['low', 'medium', 'high'])
```

Im wiÄ™ksza intensywnoÅ›Ä‡ przeÅ¼ywanego stresu tym wiÄ™ksze ryzyko pojawienia siÄ™ powaÅ¼nej depresji.

```python
df_grouped2 = df.groupby('study_load')[['study_load', 'future_career_concerns', 'anxiety_level', 'stress_level', 'depression']].median()
```

```python
df_grouped2.head(10)
```

```python
fig = px.scatter_3d(df_grouped2, x='study_load', y='future_career_concerns', z='stress_level',
                    title='WpÅ‚yw przeciÄ…Å¼enia naukÄ… i obaw o przyszÅ‚Ä… karierÄ™ zawodowÄ… na poziom stresu',
                    size_max=6, color='stress_level', opacity=0.7, color_continuous_scale='Viridis')

fig.update_layout(autosize=False, width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
